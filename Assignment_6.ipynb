```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_set = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv')
```

```python
data_set.head() #Prints first five rows.
```

```python
sns.countplot(data_set['RainToday'])
```

```python
sns.barplot(x=data_set['MinTemp'], y = data_set['MaxTemp'])
```

```python
sns.distplot(data_set['Rainfall'], bins= 20)
```

```python
sns.kdeplot(data_set['Pressure9am'], shade=True)
```

```python
sns.distplot(data_set['Temp3pm'], hist=False, rug=True)
```

```python
sns.jointplot(x=data_set['Sunshine'], y=data_set['WindGustSpeed'])
```

```python
sns.jointplot(x=data_set['Cloud9am'], y=data_set['Cloud3pm'], kind='kde')
```

```python
sns.jointplot(x=data_set['Temp9am'], y=data_set['Temp3pm'], kind='hex')
```

```python
sns.swarmplot(x=data_set['Pressure9am'], y=data_set['Pressure3pm'], data=data_set)
```

```python
sns.boxplot(x=data_set['WindDir9am'], y=data_set['Humidity3pm'], data=data_set)
```

```python
sns.boxplot(x=data_set['WindDir9am'], y=data_set['Humidity3pm'], data=data_set, hue='RainToday', dodge=True)
```

```python
sns.barplot(x=data_set['Rainfall'], y=data_set['Evaporation'], hue='Location', data=data_set)
```
